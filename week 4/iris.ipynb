```python
# loading dataframe

from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
y = iris.target

print(datasets.load_iris().target_names)

datasets.load_iris(as_frame=True)["frame"].head()
```

Let's consider the Iris dataset, a widely known database that contains information on sepal length and width for three species of the Iris family of flowering plants.

Above we can see how this database, available through [scikit-learn](https://scikit-learn.org/stable/index.html), can be imported ready for use. Once loaded, we can apply supervised models such as decision trees to create a classifier to predict the species of a flower based on its characteristics.

```python
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42) # 50% training and 50% test

# Decision Tree training
tree_class = DecisionTreeClassifier()
tree_class.fit(X_train, y_train)

# Model testing
test_prediction = tree_class.predict(X_test)
print(f"Model accuracy in test set: {metrics.accuracy_score(test_prediction, y_test):0.2f}")

# Prediction visualization
_, ax = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
scatter = ax[0].scatter(X_test[:, 0], X_test[:, 1], c=y_test)
scatter = ax[1].scatter(X_test[:, 0], X_test[:, 1], c=test_prediction)
```

The code above trains and applies a classifier for the iris database. In this case, we present the training of a decision tree classifier in the first 10 lines of the code. In the remainder of the code, we use a scatter plot to visualize and compare predicted classes (right) to the ground truth information (left).
